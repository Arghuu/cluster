```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/abcnews-date-text.csv',usecols=['headline_text'])
df.head(5)
```

```python
txt=df.headline_text
txt
```

```python
txt=txt.str.replace('[^a-zA-Z0-9]+',' ')
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
stemmer=SnowballStemmer('english')
tk=TweetTokenizer()
txt=txt.apply(lambda x:[stemmer.stem(token.lower())for token in tk.tokenize(x)]).apply (lambda token:' '.join(token))
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
txt=txt.apply(lambda x:[token for token in tk.tokenize(x) if token not in sw]).apply(lambda token:''.join(token))
txt
```

```python
txt=txt.apply(lambda x:[token for token in tk.tokenize(x) if len(token)>2]).apply (lambda token:' '.join(token))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
X=vec.fit_transform(txt)
print(X)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  Km=KMeans(n_clusters=i,random_state=1)
  Km.fit(X)
  wcss.append(Km.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('no. of clusters')
plt.ylabel('WCSS')
plt.title('ELBOW METHOD')
plt.show()
```

```python
k=KMeans(n_clusters=2,init='k-means++',random_state=1)
y=Km.fit_predict(X)
y
```

```python
df['cluster']=y
```

```python
df_0=df.loc[df['cluster']==0]
df_0.head(5)
```

```python
import wordcloud
from wordcloud import  WordCloud
text=str(df_0.headline_text)
wordcloud=WordCloud(width=800,height=400,background_color='black').generate(text.upper())
plt.figure(figsize=(500,5))
plt.imshow(wordcloud,interpolation='bicubic')
plt.axis('off')
plt.show()
```
